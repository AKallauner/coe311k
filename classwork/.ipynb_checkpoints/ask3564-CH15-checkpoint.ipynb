```python
import numpy as np
import matplotlib.pyplot as plt

def y_exact(x):
    return np.exp(-2 * x)

step_sizes = np.arange(0.01, 1.05, 0.05)
plt.figure(figsize=(10,6))

h_ref = step_sizes[0]
num_steps_ref = int(2 / h_ref)
x_ref = np.linspace(0, 2, num_steps_ref + 1)
y_ref = np.zeros(num_steps_ref + 1)
y_ref[0] = 1
for i in range(num_steps_ref):
    y_ref[i+1] = y_ref[i] + h_ref * (-2 * y_ref[i])

for h in step_sizes:
    num_steps = int(2 / h)
    x_values = np.linspace(0, 2, num_steps + 1)
    y_values = np.zeros(num_steps + 1)
    y_values[0] = 1
    for i in range(num_steps):
        y_values[i + 1] = y_values[i] + h * (-2 * y_values[i])

    y_ref_interp = np.interp(x_values, x_ref, y_ref)
    point_error = np.abs(y_values - y_ref_interp)
    total_error = np.sum(point_error)

    print(f"h = {h:.2f}, total error = {total_error:.6f}")

    plt.plot(x_values, y_values, label=f"h={h:.2f}")

x_fine = np.linspace(0, 2, 1000)
plt.plot(x_fine, y_exact(x_fine), 'k--', label='Exact')
plt.xlabel('x')
plt.ylabel('y')
plt.title("Euler's Method Approximation of dy/dx = -2y")
plt.legend(bbox_to_anchor=(1.05,1), loc='upper left')
plt.tight_layout()
plt.show()
```
