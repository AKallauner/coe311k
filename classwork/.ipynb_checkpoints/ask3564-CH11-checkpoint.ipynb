```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline
```

```python
# Known data points
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 1, 0, 1, 0, 1])
```

```python
cs = CubicSpline(x, y, bc_type='natural')

x_interp = np.linspace(0, 5, 100)
y_spline = cs(x_interp)
```

```python
coeffs = np.polyfit(x, y, 3)
poly = np.poly1d(coeffs)

y_poly = poly(x_interp)
```

```python
plt.scatter(x, y)

plt.plot(x_interp, y_poly)
plt.plot(x_interp, y_spline)

plt.xlabel("x")
plt.ylabel("y")
plt.title("Polynomial Fit vs Cubic Spline")

plt.show()
```

```python
def mse(y_true, y_pred):
    return np.mean((y_true - y_pred)**2)

poly_pred = poly(x)
spline_pred = cs(x)

print("Polynomial MSE:", mse(y, poly_pred))
print("Cubic Spline MSE:", mse(y, spline_pred))
```
