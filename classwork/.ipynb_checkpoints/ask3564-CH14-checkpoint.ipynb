```python
import numpy as np
import matplotlib.pyplot as plt

f = np.sin

a = 0
b = 2*np.pi
N = 100

x = np.linspace(a, b, N)

h = (b - a) / (N - 1)

# True derivative
true = np.cos(x)

# Forward Difference
forward = (f(x + h) - f(x)) / h

# Backward Difference
backward = (f(x) - f(x - h)) / h

# Central Difference
central = (f(x + h) - f(x - h)) / (2 * h)

# Plot
plt.subplot(2,2,1)
plt.plot(x, true)
plt.title("cos(x)")

plt.subplot(2,2,2)
plt.plot(x, forward)
plt.title("Forward Difference")

plt.subplot(2,2,3)
plt.plot(x, backward)
plt.title("Backward Difference")

plt.subplot(2,2,4)
plt.plot(x, central)
plt.title("Central Difference")

plt.show()
```
