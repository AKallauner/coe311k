```python
import numpy as np
import time
```

```python
def jacobi(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    start = time.perf_counter()

    for k in range(max_iterations):
        x_new = np.zeros_like(x)

        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]

        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            end = time.perf_counter()
            return x_new, k + 1, end - start, True

        x = x_new

    end = time.perf_counter()
    return x, max_iterations, end - start, False


def gauss_seidel(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    start = time.perf_counter()

    for k in range(max_iterations):
        x_new = x.copy()

        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i))
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n))
            x_new[i] = (b[i] - s1 - s2) / A[i][i]

        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            end = time.perf_counter()
            return x_new, k + 1, end - start, True

        x = x_new

    end = time.perf_counter()
    return x, max_iterations, end - start, False
```

```python
A = np.array([
    [0.582745, 0.48    , 0.10   , 0.     , 0.     ],
    [0.48    , 1.044129, 0.46   , 0.10   , 0.     ],
    [0.10    , 0.46    , 1.10431, 0.44   , 0.10   ],
    [0.      , 0.10    , 0.44   , 0.963889, 0.42  ],
    [0.      , 0.      , 0.10   , 0.42   , 0.522565]
], dtype=float)

b = np.array([1.162745, 2.084129, 2.20431, 1.923889, 1.042565], dtype=float)

x0 = np.zeros_like(b)
tol = 1e-6
max_iterations = 10000

jac_sol, jac_iters, jac_time, jac_conv = jacobi(A, b, x0, tol, max_iterations)
gs_sol, gs_iters, gs_time, gs_conv = gauss_seidel(A, b, x0, tol, max_iterations)

print("Jacobi:")
print("Converged:", jac_conv)
print("Iterations:", jac_iters)
print("Time (seconds):", jac_time)
print("Solution:", jac_sol)

print("\nGauss-Seidel:")
print("Converged:", gs_conv)
print("Iterations:", gs_iters)
print("Time (seconds):", gs_time)
print("Solution:", gs_sol)
```

```python
import numpy as np

A = np.array([[1, 1],
              [1, 1.0001]], dtype=float)

b = np.array([1, 1], dtype=float)

x0 = np.zeros_like(b)
tol = 1e-6
max_iterations = 100000

jac_sol, jac_iters, jac_time, jac_conv = jacobi(A, b, x0, tol, max_iterations)
gs_sol, gs_iters, gs_time, gs_conv = gauss_seidel(A, b, x0, tol, max_iterations)

print("Jacobi:")
print("Converged:", jac_conv)
print("Iterations:", jac_iters)
print("Time (seconds):", jac_time)
print("Solution:", jac_sol)

print("\nGauss-Seidel:")
print("Converged:", gs_conv)
print("Iterations:", gs_iters)
print("Time (seconds):", gs_time)
print("Solution:", gs_sol)

print("\nResidual check (Jacobi) max |Ax-b|:", np.max(np.abs(A @ jac_sol - b)))
print("Residual check (GS) max |Ax-b|:", np.max(np.abs(A @ gs_sol - b)))
```

```python
import numpy as np

def is_diagonally_dominant(A):
    n = len(A)

    for i in range(n):
        diagonal = abs(A[i][i])
        row_sum = 0

        for j in range(n):
            if j != i:
                row_sum = row_sum + abs(A[i][j])

        if diagonal < row_sum:
            return False

    return True
```

```python
A = np.array([[1, 1],
              [1, 1.0001]], dtype=float)

print(A)
print("Is the matrix diagonally dominant?", is_diagonally_dominant(A))
```
