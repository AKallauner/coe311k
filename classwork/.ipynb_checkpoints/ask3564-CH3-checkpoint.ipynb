```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def monte_carlo_pi(n):
    inside = 0

    for i in range(n):
        x = np.random.uniform(-1, 1)
        y = np.random.uniform(-1, 1)

        if x*x + y*y <= 1:
            inside = inside + 1

    pi_est = 4 * inside / n
    return pi_est
```

```python
pi_true = np.pi

absolute_errors = []
relative_errors = []

for i in range(100):
    pi_est = monte_carlo_pi(1000)

    abs_error = abs(pi_est - pi_true)
    rel_error = abs((pi_est - pi_true) / pi_true)

    absolute_errors.append(abs_error)
    relative_errors.append(rel_error)
```

```python
runs = range(1, 101)

plt.plot(runs, absolute_errors)
plt.plot(runs, relative_errors)

plt.xlabel("Run Number")
plt.ylabel("Error")
plt.title("Monte Carlo Pi Errors (100 runs, 1000 samples)")
plt.legend(["Absolute Error", "Relative Error"])
plt.show()
```
