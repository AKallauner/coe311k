```python
import numpy as np

pi_true = np.pi

sum_val = 0.0
for n in range(5000):
    term = ((-1) ** n) / (2*n + 1)
    sum_val = sum_val + term
    
    pi_est = 4 * sum_val
    
    if (n+1) % 100 == 0:
        rel_error = abs((pi_true - pi_est) / pi_true)
        print("Terms:", n + 1, " Relative error:", rel_error)
```
