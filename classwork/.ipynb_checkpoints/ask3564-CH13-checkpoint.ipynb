```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
def naive_matmul(A, B):
    n = len(A)
    C = np.zeros((n, n))

    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]

    return C
```

```python
sizes = []
n = 3

for i in range(8):
    sizes.append(n)
    n = n * 2
```

```python
naive_times = []
numpy_times = []

for size in sizes:
    A = np.random.rand(size, size)
    B = np.random.rand(size, size)

    start = time.perf_counter()
    naive_matmul(A, B)
    end = time.perf_counter()
    naive_times.append(end - start)

    start = time.perf_counter()
    np.dot(A, B)
    end = time.perf_counter()
    numpy_times.append(end - start)

    print("Size:", size, "Naive:", naive_times[-1], "NumPy:", numpy_times[-1])
```

```python
plt.plot(sizes, naive_times)
plt.plot(sizes, numpy_times)

plt.xlabel("Matrix Size")
plt.ylabel("Runtime (seconds)")
plt.title("Naive vs NumPy Matrix Multiplication")

plt.show()
```
