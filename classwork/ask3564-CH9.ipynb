```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([1, 2, 3, 4, 5])
Y = np.array([1.2, 2.8, 3.6, 4.5, 5.1])

A = np.vstack([X, np.ones(len(X))]).T
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]

print(f"Slope: {m}, Intercept: {c}")

Y_fit = m * X + c

plt.scatter(X, Y)
plt.plot(X, Y_fit)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Part 1: Data and Least Squares Best-Fit Line")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([1, 2, 3, 4, 5])
Y = np.array([1.2, 2.8, 3.6, 4.5, 5.1])

Y_rand = Y.copy()
np.random.shuffle(Y_rand)

A = np.vstack([X, np.ones(len(X))]).T
m, c = np.linalg.lstsq(A, Y_rand, rcond=None)[0]

print(f"Randomized Y -> Slope: {m}, Intercept: {c}")

Y_fit = m * X + c

plt.scatter(X, Y_rand)
plt.plot(X, Y_fit)
plt.xlabel("X")
plt.ylabel("Y (shuffled)")
plt.title("Part 2: Randomized Data and Best-Fit Line")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

m_true = 2.5
c_true = -0.4

X = np.arange(0, 1.01, 0.01)

noise = np.random.uniform(-0.05, 0.05, len(X))
Y = m_true * X + c_true + noise

A = np.vstack([X, np.ones(len(X))]).T
m_fit, c_fit = np.linalg.lstsq(A, Y, rcond=None)[0]

print(f"True line: m={m_true}, c={c_true}")
print(f"Fit line:  m={m_fit}, c={c_fit}")

Y_fit = m_fit * X + c_fit
Y_true = m_true * X + c_true

plt.scatter(X, Y, s=10)
plt.plot(X, Y_true)
plt.plot(X, Y_fit)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Original Line vs Best Fit Line")
plt.show()
```
