```python
import numpy as np
import matplotlib.pyplot as plt

k = 1.0
x_start = 0
x_end = 10
y_0 = 1.0
v_0 = 0.0

h_values = [0.01, 0.05, 0.1, 0.2, 0.3]

plt.figure()

for h in h_values:
    # Coefficient matrix for Euler's method
    A = np.array([[1, h],
                  [-h*k, 1]])

    eigvals = np.linalg.eigvals(A)

    print("h =", h)
    print("Eigenvalues:", eigvals)

    if np.all(np.abs(eigvals) < 1):
        print("Solution is Stable")
    else:
        print("Solution may not be Entirely Stable ;)")

    print()

    # Simulation
    x_values = np.arange(x_start, x_end + h, h)
    y_values = np.zeros(len(x_values))
    v_values = np.zeros(len(x_values))

    y_values[0] = y_0
    v_values[0] = v_0

    for i in range(1, len(x_values)):
        y_values[i] = y_values[i-1] + h * v_values[i-1]
        v_values[i] = v_values[i-1] + h * (-k * y_values[i-1])

    plt.plot(x_values, y_values, label=f"h={h}")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Euler Stability Analysis")
plt.legend()

plt.show()
```
