```python
import numpy as np
import matplotlib.pyplot as plt

k = 1.0
x_start = 0
x_end = 10
y_0 = 1.0
v_0 = 0.0

h_values = [0.01, 0.02, 0.03, 0.05, 0.1, 0.2, 0.3, 0.5]

plt.figure()

for h in h_values:
    x_values = np.arange(x_start, x_end + h, h)
    y_values = np.zeros(len(x_values))
    v_values = np.zeros(len(x_values))

    y_values[0] = y_0
    v_values[0] = v_0

    for i in range(1, len(x_values)):
        y_values[i] = y_values[i-1] + h * v_values[i-1]
        v_values[i] = v_values[i-1] + h * (-k * y_values[i-1])

    plt.plot(x_values, y_values, label=f"h={h}")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Euler Method with Different Step Sizes")
plt.legend()

plt.show()
```

The solution starts to become unstable around h = 0.2 and up. As the step sizes increase, the solution grows instead of smoothly oscillating, which shows instability. Increasing k makes the system oscillate faster, and causes instability at the smaller step sizes.

```python
k = 10
h = 0.01

x_values = np.arange(x_start, x_end + h, h)
y_values = np.zeros(len(x_values))
v_values = np.zeros(len(x_values))

y_values[0] = y_0
v_values[0] = v_0

for i in range(1, len(x_values)):
    y_values[i] = y_values[i-1] + h * v_values[i-1]
    v_values[i] = v_values[i-1] + h * (-k * y_values[i-1])

plt.plot(x_values, y_values)
plt.title("k = 10, h = 0.01")
plt.show()
```

Euler's method doesn't perform very well for larger step sizes. As h increases, the error becomes large and the solution becomes unstable. You would need smaller step sizes for more accurate results.
