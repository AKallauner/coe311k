```python
import numpy as np
import time
```

```python
def jacobi(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    start = time.perf_counter()

    for k in range(max_iterations):
        x_new = np.zeros_like(x)

        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]

        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            end = time.perf_counter()
            return x_new, k + 1, end - start, True

        x = x_new

    end = time.perf_counter()
    return x, max_iterations, end - start, False


def gauss_seidel(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    start = time.perf_counter()

    for k in range(max_iterations):
        x_new = x.copy()

        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i))
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n))
            x_new[i] = (b[i] - s1 - s2) / A[i][i]

        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            end = time.perf_counter()
            return x_new, k + 1, end - start, True

        x = x_new

    end = time.perf_counter()
    return x, max_iterations, end - start, False
```

```python
A = np.array([
    [10, -1,  2,  0,  0],
    [-1, 11, -1,  3,  0],
    [ 2, -1, 10, -1,  0],
    [ 0,  3, -1,  8, -2],
    [ 0,  0,  0, -2,  9]
], dtype=float)

b = np.array([14, 30, 26, 25, 37], dtype=float)

x0 = np.zeros_like(b)
tol = 1e-6
max_iterations = 10000

jac_sol, jac_iters, jac_time, jac_conv = jacobi(A, b, x0, tol, max_iterations)
gs_sol, gs_iters, gs_time, gs_conv = gauss_seidel(A, b, x0, tol, max_iterations)

print("Jacobi:")
print("Converged:", jac_conv)
print("Iterations:", jac_iters)
print("Time (seconds):", jac_time)
print("Solution:", jac_sol)

print("\nGauss-Seidel:")
print("Converged:", gs_conv)
print("Iterations:", gs_iters)
print("Time (seconds):", gs_time)
print("Solution:", gs_sol)
```

For 6a, both methods converged with a tolerance of 10^-6. Gauss Seidel needed less iterations and runtime than Jacobi. This is beacause Gauss-Seidel usues updated values for each iteration, which improves its convergence speed.

```python
A = np.array([
    [1, 2, 3, 0, 0],
    [2, 1, 2, 3, 0],
    [3, 2, 1, 2, 3],
    [0, 3, 2, 1, 2],
    [0, 0, 3, 2, 1]
], dtype=float)

b = np.array([14, 22, 33, 26, 22], dtype=float)

x0 = np.zeros_like(b)
tol = 1e-6
max_iterations = 10000

jac_sol, jac_iters, jac_time, jac_conv = jacobi(A, b, x0, tol, max_iterations)
gs_sol, gs_iters, gs_time, gs_conv = gauss_seidel(A, b, x0, tol, max_iterations)

print("Jacobi (6b):")
print("Converged:", jac_conv)
print("Iterations:", jac_iters)
print("Time (seconds):", jac_time)

print("\nGauss-Seidel (6b):")
print("Converged:", gs_conv)
print("Iterations:", gs_iters)
print("Time (seconds):", gs_time)
```
