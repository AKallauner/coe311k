```python
import math

def calc_hyp(a, b):
    h = math.sqrt(a*a + b*b)
    return h
```

```python
my_h = calc_hyp(3, 4)
print(my_h)
```

```python
import random
def montecarlo_pi (number_of_shots):
    counter = 0
    for i in range(0,number_of_shots):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        
        if (x**x + y*y < 1):
            number_of_hits = number_of_hits + 1
            
    return 4*number_of_hits/number_of_shots
```
