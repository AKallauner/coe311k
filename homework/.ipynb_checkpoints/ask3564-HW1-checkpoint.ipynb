```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# asked chatGPT to help me understand and write Monte Carlo Pi since I missed class on Thursday
def monte_carlo_pi(n_samples):
    inside = 0
    for _ in range(n_samples):
        x = np.random.uniform(-1, 1)
        y = np.random.uniform(-1, 1)
        if x*x + y*y <= 1:
            inside += 1
    return 4 * inside / n_samples
```

```python
pi_true = np.pi
samples_list = list(range(10, 1001, 20))
errors = []

for n in samples_list:
    pi_est = monte_carlo_pi(n)
    errors.append(abs(pi_est - pi_true))
```

```python
plt.plot(samples_list, errors)
plt.xlabel("Number of samples")
plt.ylabel("Absolute Error")
plt.title("Monte Carlo Pi Approximation Error")
plt.show()
```
