```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return -2*y*np.cos(x)

def exact(x):
    return np.exp(-2*np.sin(x))

h_values = [0.01, 0.05, 0.15, 0.2, 0.25, 0.5, 1]

x_start = 0
x_end = 10*np.pi
y0 = 1

avg_errors = []
local_errors = []

plt.figure()

for h in h_values:
    x = np.arange(x_start, x_end + h, h)
    y = np.zeros(len(x))
    y[0] = y0

    for i in range(1, len(x)):
        y[i] = y[i-1] + h * f(x[i-1], y[i-1])

    y_true = exact(x)

    # Plot Euler solutions
    plt.plot(x, y, label=f"h={h}")

    # Average error
    avg_error = np.mean(np.abs(y - y_true))
    avg_errors.append(avg_error)

    # Local error near x ≈ 15
    idx = np.argmin(np.abs(x - 15))
    local_error = abs(y[idx] - exact(x[idx]))
    local_errors.append(local_error)

plt.xlabel("x")
plt.ylabel("y")
plt.title("Euler Method for Different h")
plt.legend(loc="upper right")
plt.show()

# Plot average error vs h
plt.plot(h_values, avg_errors, 'o-')
plt.xlabel("h")
plt.ylabel("Average Error")
plt.title("Average Error vs Step Size")
plt.show()

# Plot local error vs h
plt.plot(h_values, local_errors, 'o-')
plt.xlabel("h")
plt.ylabel("Local Error at x ≈ 15")
plt.title("Local Error vs Step Size")
plt.show()
```

As the step size increases, the error in Euler's method also increases with it. Smaller step sizes produce more accurate results since the approximation follows along the curve more closely. The average error grows with larger step size, and the local error at x=15 also increases. This shows that Euler's method is sensitive to step size, and larger steps tend to produce less accurate solutions.
