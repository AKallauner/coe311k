```python
import numpy as np

h_values = [0.01, 0.05, 0.15, 0.2, 0.25, 0.5, 1]

for h in h_values:
    val = abs(1 + h*(-2))
    print("h =", h, "|1 + hλ| =", val)
```

The equation can be written as y' = λ(x)y where λ(x) = -2cos(x). Since cos(x) is between -1 and 1, λ(x) is between -2 and 2. Using the stability condition |1 + hλ| < 1 and the worst case λ = -2, we get |1 - 2h| < 1, which gives 0 < h < 1. This means the step size must be less than 1 for stability. Because λ depends on x, it changes over time but using the bounds of cos(x) allows us to determine a stability condition. Larger step sizes can cause instability.
