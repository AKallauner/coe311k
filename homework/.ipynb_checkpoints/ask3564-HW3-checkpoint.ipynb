```python
import numpy as np
import matplotlib.pyplot as plt

x_vals = np.arange(0, 10.1, 0.1)
```

```python
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result = result * i
    return result
```

```python
def taylor_sin(x, n_terms):
    total = 0.0

    for k in range(n_terms):
        power = 2*k + 1
        sign = (-1) ** k
        total = total + sign * (x ** power) / factorial(power)

    return total
```

```python
sin_exact = np.sin(x_vals)

errors_1 = []
errors_2 = []
errors_3 = []
errors_4 = []

for x in x_vals:
    approx1 = taylor_sin(x, 1)
    approx2 = taylor_sin(x, 2)
    approx3 = taylor_sin(x, 3)
    approx4 = taylor_sin(x, 4)

    exact = np.sin(x)

    errors_1.append(abs(exact - approx1))
    errors_2.append(abs(exact - approx2))
    errors_3.append(abs(exact - approx3))
    errors_4.append(abs(exact - approx4))
```

```python
plt.plot(x_vals, errors_1, label="1 term")
plt.plot(x_vals, errors_2, label="2 terms")
plt.plot(x_vals, errors_3, label="3 terms")
plt.plot(x_vals, errors_4, label="4 terms")

plt.xlabel("x")
plt.ylabel("Absolute Error")
plt.title("Taylor Series Truncation Error for sin(x)")
plt.legend()
plt.show()
```

```python
# As more Taylor series get included, the truncation error decreases since the polynomial can approximate the sin function better. The error doesn't decrease uniformally for all values of x, and for larger values the higher powers grow quickly. WHen larger values of x get tested, the error greatly increases for small numbers of terms which shows that more terms are needed to keep accuracy farther from zero.
```
